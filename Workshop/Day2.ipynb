```python
fig , ax=plt.subplots(1,3,figsize=(15,5))

ax[0].plot([1,2,3,4],[3,4,5,6],'r')
ax[0].set_title("PLOT 1")

ax[1].plot([2,4,5,6,7],[3,4,5,7,8],'y')
ax[1].set_title("PLOT 2")

ax[2].plot([2,3,6,7],[3,4,6,7],'b')
ax[2].set_title("PLOT 3")

plt.tight_layout()
plt.show()
```

```python
# Create a 1x3 grid of subplots (1 row, 3 columns)
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

# Plot on each subplot
ax[0].plot([1, 2, 3], [1, 4, 9], 'r')  # Red line
ax[0].set_title("nandu")

ax[1].plot([1, 2, 3], [1, 2, 3], 'g')  # Green line
ax[1].set_title("Plot 2")

ax[2].plot([1, 2, 3], [9, 4, 1], 'b')  # Blue line
ax[2].set_title("Plot 3")

# Adjust layout to avoid overlap
plt.tight_layout()
plt.show()
```

```python
# Create a 1x3 grid of subplots (1 row, 3 columns)
fig, ax = plt.subplots(2, 2, figsize=(8, 6))

# Plot on each subplot
ax[0,0].plot([1, 2, 3], [1, 4, 9], 'r')  # Red line
ax[0,0].set_title("Plot 1")

ax[0,1].plot([1, 2, 3], [1, 2, 3], 'g')  # Green line
ax[0,1].set_title("Plot 2")

ax[1,0].plot([1, 2, 3], [9, 4, 1], 'b')  # Blue line
ax[1,0].set_title("Plot 3")

ax[1,1].plot([1, 2, 3], [2.4, 5, 9], 'y')  # Blue line
ax[1,1].set_title("Plot 4")


# Adjust layout to avoid overlap
plt.tight_layout()
plt.show()
```

```python
# Create a 2x2 grid of subplots
fig, ax = plt.subplots(2, 2, figsize=(8, 6))


# Plot on each subplot
ax[0, 0].plot([1, 2, 3], [1, 4, 9], 'r')  # Top-left
ax[0, 0].set_title("Top Left")

ax[0, 1].plot([1, 2, 3], [1, 2, 3], 'g')  # Top-right
ax[0, 1].set_title("Top Right")

ax[1, 0].plot([1, 2, 3], [9, 4, 1], 'b')  # Bottom-left
ax[1, 0].set_title("Bottom Left")

ax[1, 1].plot([1, 2, 3], [3, 3, 3], 'k')  # Bottom-right
ax[1, 1].set_title("Bottom Right")

# Adjust layout
plt.tight_layout()
plt.show()
```
