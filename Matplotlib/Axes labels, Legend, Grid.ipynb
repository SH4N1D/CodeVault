```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
day=np.arange(1,8)
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.scatter(day,max_t,label='max')
plt.scatter(day,min_t,label='min')
plt.scatter(day,avg_t,label='avg')
plt.xlabel('Days')
plt.ylabel('temperature')
plt.title('Weather Data')
plt.legend(loc='lower left',shadow=True,fontsize='small')
plt.show()
```

```python
plt.plot(day,max_t,label='max')
plt.plot(day,min_t,label='min')
plt.plot(day,avg_t,label='avg')
plt.xlabel('Days')
plt.ylabel('temperature')
plt.title('Weather Data')
plt.legend(loc='lower left',shadow=True,fontsize='small')
plt.grid()
```

```python
fig,ax=plt.subplots(1,3,figsize=(15,5))

ax[0].plot([1,2,3,4],[3,4,5,6],'r')
ax[0].set_title("PLOT 1")

ax[1].plot([2,4,5,6,7],[3,4,5,7,8],'y')
ax[1].set_title("PLOT 2")

ax[2].plot([2,3,6,7],[3,4,6,7],'b')
ax[2].set_title("PLOT 3")

plt.tight_layout()
plt.show()
```

```python
fig,k=plt.subplots(2,2,figsize=(10,6))

k[0,0].plot([50,51,52,48,47,49,46],'r')
k[0,0].set_title('PLOT1')
k[0,0].set_xlabel('Day1')
k[0,0].set_ylabel('Temperature')

k[0,1].plot([43,42,40,44,33,35,37],'g')
k[0,1].set_title('Plot2')

k[1,0].plot([45,48,48,46,40,42,41],'b')
k[1,0].set_title('Plot3')

k[1,1].plot([45,47,48,42,45,50,46],'y')
k[1,1].set_title('Plot4')

plt.tight_layout()
plt.show()
```
