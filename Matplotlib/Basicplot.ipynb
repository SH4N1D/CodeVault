```python
import matplotlib.pyplot as plt
```

```python
plt.plot([2,5,9,8,16,14])
plt.show()
```

```python
plt.plot([1,2,3,4,5],[1,4,9,16,25],'r--.')
plt.xlabel(" Time  ")
plt.ylabel("  Work  ")
plt.title("Job")
```

```python
plt.plot([1,2,3,4,5],[1,4,9,16,25],color="blue",marker='D',linestyle='--',markersize=3)
plt.xlabel(" Time  ")
plt.ylabel("  Work  ")
plt.title("Job")
```

alpha for opacity

```python
plt.plot([2,5,9,8,16,14],alpha=0.3)
plt.show()
```
