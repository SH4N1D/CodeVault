```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
blood_sugar=[113,85,90,150,149,88,93,115,135,80,77,82,129]
```

```python
plt.hist(blood_sugar)
plt.show()
```

```python
plt.hist(blood_sugar,bins=3)
plt.show()
```

```python
plt.hist(blood_sugar,bins=5,rwidth=0.95)
plt.show()
```

bins means no of boxes (default 10), rwidth is the relative width

```python
plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95)
plt.show()
```

```python
plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95,color='cyan')
plt.show()
```

```python
plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95,color='red',histtype='step')
plt.show()
```

```python
blood_sugar_men=[113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_women=[67,98,89,120,133,150,84,69,89,79,120,112,100]
```

```python
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.95)
plt.show()
```

```python
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.95,color=['red','cyan'],label=['men','women'])
plt.xlabel('Sugar Range')
plt.ylabel('Count')
plt.title('Blood Sugar levels')
plt.legend()
plt.show()
```

```python
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.95,color=['red','cyan'],label=['men','women'],orientation='horizontal')
plt.xlabel('Sugar Range')
plt.ylabel('Count')
plt.title('Blood Sugar levels')
plt.legend()
plt.show()
```
