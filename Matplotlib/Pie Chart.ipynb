```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
vals=[1400,600,300,410,250]
label=['Home Rent','Food','Internet Bill','Car','others']
```

```python
plt.pie(vals,labels=label)
plt.show()
```

```python
plt.pie(vals,labels=label)
plt.axis('equal')
plt.show()
```

```python
plt.pie(vals,labels=label,radius=0.8)
plt.axis('equal')
plt.show()
```

```python
plt.pie(vals,labels=label,radius=0.8,autopct='%0.0f%%')
plt.axis('equal')
plt.show()
```

```python
plt.pie(vals,labels=label,radius=0.8,autopct='%0.2f%%')
plt.axis('equal')
plt.show()
```

```python
plt.pie(vals,labels=label,radius=0.8,autopct='%0.0f%%',shadow=True)
plt.axis('equal')
plt.show()
```

```python
plt.pie(vals,labels=label,radius=0.8,autopct='%0.0f%%',explode=[0.09,0.1,0.09,0.1,0.09])
plt.axis('equal')
plt.show()
```

```python
plt.pie(vals,labels=label,radius=0.8,autopct='%0.0f%%',startangle=150,explode=[0.09,0.1,0.09,0.1,0.09])
plt.axis('equal')
plt.show()
```
