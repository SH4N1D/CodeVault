```python
import pandas as pd
```

```python
df1=pd.DataFrame({
    'city':['new york','chicago','orlando'],
    'temperature':[21,14,35]
})
df1
```

```python
df2=pd.DataFrame({
    'city':['new york','chicago','orlando'],
    'humidity':[65,68,75]
})
df2
```

```python
df3=pd.merge(df1,df2,on='city')
df3
```

```python
df1=pd.DataFrame({
    'city':['new york','chicago','orlando','baltimore'],
    'temperature':[21,14,35,31]
})
df1
```

```python
df2=pd.DataFrame({
    'city':['new york','chicago','sanfrancisco'],
    'humidity':[65,68,75]
})
df2
```

```python
df3=pd.merge(df1,df2,on='city',how='outer')
df3
```

```python
df3=pd.merge(df1,df2,on='city',how='left')
df3
```

```python
df3=pd.merge(df1,df2,on='city',how='outer',indicator=True)
df3
```
